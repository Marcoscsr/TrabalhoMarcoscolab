```
import pandas as pd
df = pd.read_csv('netflix.csv')
df.head()
```

```
cont_classificação = df['classificação'].value_counts()
plt.figure(figsize = (5, 5))
plt.bar(cont_classificação.index, cont_classificação.values, color = ['red', 'black'])
plt.title('Classificação indicativa')
plt.xlabel('Anos')
plt.ylabel('Quantidade')
plt.show()

#Qual Classificação indicativa é maior entre as séries?
#Resposta: 18+
```
